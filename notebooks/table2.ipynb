```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../scripts/sac.csv', header=None)
df.columns = ['seed', 'return', 'env', 'j', 'ASL']

ENV_LIST = ["Pendulum-v1", "LunarLanderContinuous-v2", "Hopper-v2", "Walker2d-v2", "Ant-v2", "HalfCheetah-v2", "Humanoid-v2", "InvertedPendulum-v2", "InvertedDoublePendulum-v2", "Reacher-v2", "Swimmer-v2"]
MAX_REWARDS = [1350, 300, 3600, 5000, 6500, 17000, 10500, 1000, 9400, 60, 350]
```

```python
AUC_df = pd.DataFrame(columns=['Environment', 'SAC-1', 'SAC-2', 'SAC-4', 'SAC-8', 'SAC-16'])
for loc, env_name in enumerate(ENV_LIST):
    AUC_df.loc[loc] = [env_name, 0, 0,0,0,0]
    max_reward = MAX_REWARDS[loc]

    for train_steps in [1,2,4,8,16]:
            normalized_aucs = []
            for seed in range(5):
                filtered_df = df[(df['seed'] == seed) & (df['j'] == train_steps) & (df['env'] == env_name)]

                steps_list = filtered_df['ASL'].values
                avg_rewards = filtered_df['return'].values
                max_auc = max_reward * 30
                if env_name == 'Reacher-v2':
                    avg_rewards = avg_rewards + 65
                if env_name == 'Pendulum-v1':
                    avg_rewards = avg_rewards + 1550
                    
                normalized_auc = np.trapz(avg_rewards, steps_list)/max_auc
                normalized_aucs.append(normalized_auc)
            average_normalized_auc = np.mean(normalized_aucs)
            AUC_df.loc[loc,'SAC-'+str(train_steps)] = average_normalized_auc
```
